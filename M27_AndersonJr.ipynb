Depois disso eu percebi que a quantidade de dados faltantes iam me atrapalhar, então comecei a limpar ainda mais os dados. Descobri que precisaria da coluna mês novamente depois de ler o briefing 2.2.

```python
# Filtrar os dados para CONFINS como origem ou destino
confins_origem = Dados_Estatisticos[Dados_Estatisticos['AEROPORTO_DE_ORIGEM_NOME'] == 'CONFINS']
confins_destino = Dados_Estatisticos[Dados_Estatisticos['AEROPORTO_DE_DESTINO_NOME'] == 'CONFINS']

# Concatenar os DataFrames filtrados
confins_df = pd.concat([confins_origem, confins_destino], ignore_index=True)

# Remover mais colunas ainda
confins_df.drop(columns=['AEROPORTO_DE_ORIGEM_NOME', 'AEROPORTO_DE_DESTINO_NOME', 'CORREIO_KG', 'ASK', 'RPK', 'DECOLAGENS', 'BAGAGEM_KG'], inplace=True)

# Definir o índice como 'ANO'
if 'ANO' in confins_df.columns:
    confins_df = confins_df.set_index('ANO')

# Converter o índice para int
confins_df.index = confins_df.index.astype(int)
confins_df.head()
```

Aqui eu percebi que manter o índice como ANO não ia me ajudar se eu quisesse ordenar por mês em uma Dash, aí fui refazer. Deixando aqui registrado todo o processo.

```python
# Definir 'DATA' como índice
confins_df = confins_df.set_index('DATA')

# Ordenar o DataFrame por índice
confins_df = confins_df.sort_index()

# Reamostrando dataframe
confins_df = confins_df.resample(rule='m').sum()

# Reorganizar as colunas
confins_df = confins_df.reset_index()
confins_df = confins_df[['DATA', 'PASSAGEIROS_TOTAL', 'PASSAGEIROS_PAGOS', 'PASSAGEIROS_GRATIS']]

confins_df.head()
```

```python
# Criar outro índice
confins_df['DATA'] = pd.to_datetime(confins_df['DATA'])
confins_df.set_index('DATA', inplace=True)
confins_df.head()
```

Essa parte novamente veio do 2.2 e 2.3. Modifiquei o formato do índice, criei novas colunas para obedecer ao que foi pedido, ordenei elas de forma que ficasse como pedido e criei o arquivo CSV sem o índice.

```python
# Colocar o índice DATA no formato certo
confins_final['DATA'] = pd.to_datetime(confins_final.index)

# Adicionar colunas 'ANO' e 'MES'
confins_final['ANO'] = confins_final['DATA'].dt.year
confins_final['MES'] = confins_final['DATA'].dt.month

# Adicionar coluna 'ANO-MES' no formato desejado
confins_final['ANO-MES'] = confins_final['DATA'].dt.strftime('%Y-%m')

# Reordenar as colunas
colunas_reordenadas = ['ANO-MES', 'ANO', 'MES', 'PASSAGEIROS_PAGOS', 'PASSAGEIROS_GRATIS', 'PASSAGEIROS_TOTAL']
confins_final = confins_final[colunas_reordenadas]

# Salvar o DataFrame em um arquivo CSV sem o índice
confins_final.to_csv('flights.csv', index=False)
```
